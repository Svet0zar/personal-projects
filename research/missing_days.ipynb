```python
import modules.get_ohlcv_features as gof

ohlcv_feats = gof.make_ohlcv_features_lgbm(df_ohlcv)
ohlcv_feats
```

```python
import matplotlib.pyplot as plt

# Find missing business days in ohlcv_feats index
all_bdays = pd.date_range(ohlcv_feats.index.min(), ohlcv_feats.index.max(), freq="B")
missing_dates = all_bdays.difference(ohlcv_feats.index)

print(f"Number of missing business days: {len(missing_dates)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

s = pd.Series(1, index=missing_dates).reindex(all_bdays, fill_value=0)
df = pd.DataFrame({"missing": s.astype(bool)})

monthly_rate = df["missing"].groupby([df.index.year, df.index.month]).mean()
heat = monthly_rate.unstack(fill_value=0).sort_index()

fig, ax = plt.subplots(figsize=(12, 6))
im = ax.imshow(heat.values, aspect='auto', interpolation='nearest')

ax.set_xticks(np.arange(12))
ax.set_xticklabels(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
ax.set_yticks(np.arange(len(heat.index)))
ax.set_yticklabels(heat.index)
ax.set_xlabel("Month")
ax.set_ylabel("Year")
ax.set_title("Share of Missing Business Days by Month")
cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Share missing (0–1)")
plt.tight_layout()
plt.show()
```
