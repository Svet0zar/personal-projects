```python
res, smoothed_probs = gr.fit_mrs(df["Matif_Prices"], k_regimes=3)

print(res.summary())

# Inspect regime probabilities
smoothed_probs.plot(figsize=(12,6), title="Smoothed Regime Probabilities")

# Assign most likely regime at each date
regime = smoothed_probs.idxmax(axis=1)
regime.name = "regime_mrs"

print(regime.value_counts(normalize=True))

def enforce_persistence(regime_series, min_duration=20):
    regime = regime_series.copy()
    last = regime.iloc[0]
    count = 1
    for i in range(1, len(regime)):
        if regime.iloc[i] == last:
            count += 1
        else:
            if count < min_duration:
                # overwrite the short blip with previous regime
                regime.iloc[i-count:i] = last
            last = regime.iloc[i]
            count = 1
    return regime

# apply
reg_clean = enforce_persistence(regime, min_duration=20)

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def plot_price_with_multi_regimes(
    price: pd.Series,
    regime: pd.Series,                 # can be ints (0/1/2) or strings
    title: str = "Price with Regimes",
    colors: dict | None = None,        # e.g., {0:"#6ea8fe",1:"#ffd166",2:"#ef476f"}
    alpha: float = 0.18,
    shift_one: bool = True,            # use regime(t-1) to avoid look-ahead on t
    figsize=(18, 6),
):
    """
    Shade the background by regime (works with 2+ regimes).

    - price: price series with DatetimeIndex
    - regime: regime labels indexed by date (can be returns index; we align)
    - colors: optional mapping from regime value -> color
    - shift_one: if True, regime is shifted by 1 day so today's regime uses info up to t-1
    """
    # 1) Align and clean
    price = price.dropna()
    reg = regime.reindex(price.index).ffill()
    if shift_one:
        reg = reg.shift(1).ffill()
    # cast to a stable type for equality tests
    if pd.api.types.is_categorical_dtype(reg):
        reg = reg.astype(reg.dtype)  # keep categories
    else:
        # if ints-as-floats slipped in, fix them
        try:
            reg = reg.astype(int)
        except Exception:
            pass

    # 2) Color map
    uniq = pd.Index(pd.unique(reg.dropna()))
    if colors is None:
        # default palette
        base = ["#6ea8fe", "#ffd166", "#ef476f", "#7bd389", "#b084cc", "#8ecae6"]
        colors = {k: base[i % len(base)] for i, k in enumerate(uniq)}
    # ensure every regime has a color
    for k in uniq:
        colors.setdefault(k, "#dddddd")

    # 3) Build contiguous blocks
    vals = reg.to_numpy()
    idx = price.index
    change_ix = np.r_[0, 1 + np.where(vals[1:] != vals[:-1])[0], len(vals)]

    # 4) Plot
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot(idx, price.values, color="black", lw=1.2, label="Price")

    for i in range(len(change_ix) - 1):
        s, e = change_ix[i], change_ix[i+1]
        r = vals[s]
        if pd.isna(r):
            continue
        ax.axvspan(idx[s], idx[e-1], color=colors.get(r, "#dddddd"), alpha=alpha, lw=0)

    # legend
    handles = [plt.Line2D([0],[0], color=c, lw=8, alpha=alpha) for c in colors.values()]
    labels  = [f"Regime {k}" for k in colors.keys()]
    ax.legend([plt.Line2D([0],[0], color="black", lw=1.2, label="Price"), *handles],
              ["Price", *labels], loc="upper left", frameon=True)

    ax.set_title(title)
    ax.set_ylabel("Price")
    fig.tight_layout()
    plt.show()

# plot_price_with_multi_regimes(df["Matif_Prices"], REGIME,
#                         title="Matif Wheat Price with Realized Vol Regimes")
# regime_mrs is the argmax of smoothed probs (values 0/1/2), indexed by returns
plot_price_with_multi_regimes(
    price=df["Matif_Prices"],
    regime=reg_clean,  # the series you computed
    title="Matif Wheat Price with 3-State MRS Regimes",
    colors={0:"#6ea8fe", 1:"#ffd166", 2:"#ef476f"},  # optional
    alpha=0.6,
    shift_one=True,     # avoids look-ahead
)

```

```python
import numpy as np
import pandas as pd
from typing import Tuple, List, Dict, Optional

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from scipy.stats import spearmanr

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

# ---------- Target: H-day cumulative log return ----------
def make_horizon_logret_target(price: pd.Series, H: int = 10) -> pd.Series:
    """y_t = sum of next H daily log returns (shifted so label is at t)."""
    r = np.log(price / price.shift(1))
    y = r.rolling(H).sum().shift(-H)
    return y

# ---------- Purge/embargo masking ----------
def biz_offset_pos(pos: int, days: int, n: int) -> int:
    return int(np.clip(pos + days, 0, n))

def between_mask(idx: pd.DatetimeIndex, start, end):
    return (idx >= start) & (idx <= end)

# ---------- Train-only numeric scaler/imputer ----------
def fit_num_imputer_scaler(X_train_num: pd.DataFrame):
    imp = SimpleImputer(strategy="median")
    sc  = StandardScaler()
    Xn  = imp.fit_transform(X_train_num)
    Xs  = sc.fit_transform(Xn)
    return imp, sc

def transform_num_with(imp, sc, X_num: pd.DataFrame):
    return sc.transform(imp.transform(X_num))

# ---------- Train-fitted one-hot for categoricals ----------
def one_hot_train_test(
    X_tr_cat: pd.DataFrame,
    X_te_cat: pd.DataFrame
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Impute categoricals by most-frequent on TRAIN, fix categories from train,
    then one-hot both, aligned to train columns.
    """
    if X_tr_cat.shape[1] == 0:
        return (pd.DataFrame(index=X_tr_cat.index), pd.DataFrame(index=X_te_cat.index))

    imp = SimpleImputer(strategy="most_frequent")
    Xtr_imp = pd.DataFrame(imp.fit_transform(X_tr_cat), index=X_tr_cat.index, columns=X_tr_cat.columns)
    Xte_imp = pd.DataFrame(imp.transform(X_te_cat),   index=X_te_cat.index, columns=X_te_cat.columns)

    # lock categories from train
    for c in Xtr_imp.columns:
        cats = pd.Categorical(Xtr_imp[c]).categories
        Xtr_imp[c] = pd.Categorical(Xtr_imp[c], categories=cats)
        Xte_imp[c] = pd.Categorical(Xte_imp[c], categories=cats)

    Xtr_oh = pd.get_dummies(Xtr_imp, drop_first=False)
    Xte_oh = pd.get_dummies(Xte_imp, drop_first=False)

    # align to train one-hot columns
    Xte_oh = Xte_oh.reindex(columns=Xtr_oh.columns, fill_value=0)
    return Xtr_oh, Xte_oh

# ---------- Sequence maker (no leakage) ----------
def make_sequences_from_slice(
    X: pd.DataFrame,
    y: pd.Series,
    start_ts, end_ts,
    seq_len: int = 60
):
    """Build sequences [T,F] ending at t, with label y[t], for t in [start_ts, end_ts]."""
    Xs, ys, out_idx = [], [], []
    idx = X.index
    # valid label times
    valid = y.index[(y.index >= start_ts) & (y.index <= end_ts) & y.notna()]
    for t in valid:
        pos = idx.get_indexer([t])[0]
        if pos - (seq_len - 1) < 0:
            continue
        win_start = idx[pos - (seq_len - 1)]
        if win_start < start_ts:  # window must be inside the slice
            continue
        x_win = X.iloc[pos - (seq_len - 1): pos + 1].values  # [T,F]
        if np.isnan(x_win).any():  # after impute/encode/scale, there shouldn't be NaN; skip if any
            continue
        Xs.append(x_win)
        ys.append(y.loc[t])
        out_idx.append(t)
    if not Xs:
        return np.empty((0, seq_len, X.shape[1])), np.array([]), pd.DatetimeIndex([])
    return np.stack(Xs), np.asarray(ys), pd.DatetimeIndex(out_idx)

# ---------- Torch dataset ----------
class SeqDataset(Dataset):
    def __init__(self, X: np.ndarray, y: np.ndarray):
        self.X = torch.as_tensor(X, dtype=torch.float32)
        self.y = torch.as_tensor(y, dtype=torch.float32)
    def __len__(self):
        return self.X.shape[0]
    def __getitem__(self, i):
        return self.X[i], self.y[i]
```

```python
class LSTMRegressor(nn.Module):
    def __init__(self, n_features: int, hidden: int = 64, layers: int = 1, dropout: float = 0.1):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=n_features,
            hidden_size=hidden,
            num_layers=layers,
            batch_first=True,
            dropout=dropout if layers > 1 else 0.0
        )
        self.head = nn.Sequential(
            nn.Linear(hidden, 64),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(64, 1)
        )
    def forward(self, x):  # x: [B,T,F]
        out, _ = self.lstm(x)        # [B,T,H]
        last = out[:, -1, :]         # [B,H]
        return self.head(last).squeeze(-1)  # [B]

@torch.no_grad()
def eval_model(model, loader, device):
    model.eval()
    preds, trues = [], []
    for xb, yb in loader:
        xb = xb.to(device)
        pred = model(xb).cpu().numpy()
        preds.append(pred)
        trues.append(yb.numpy())
    yhat = np.concatenate(preds) if preds else np.array([])
    yt   = np.concatenate(trues) if trues else np.array([])
    return yhat, yt

def train_lstm_one_fold(
    X_tr_seq: np.ndarray, y_tr: np.ndarray,
    X_val_seq: np.ndarray, y_val: np.ndarray,
    hidden: int = 64, layers: int = 1, dropout: float = 0.1,
    lr: float = 1e-3, weight_decay: float = 1e-4,
    batch_size: int = 128, max_epochs: int = 200,
    early_stopping_patience: int = 20,
    device: str = "cuda" if torch.cuda.is_available() else "cpu"
):
    n_features = X_tr_seq.shape[2]
    model = LSTMRegressor(n_features, hidden, layers, dropout).to(device)
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    loss_fn = nn.SmoothL1Loss()  # Huber
    train_loader = DataLoader(SeqDataset(X_tr_seq, y_tr), batch_size=batch_size, shuffle=True, drop_last=False)
    val_loader   = DataLoader(SeqDataset(X_val_seq, y_val), batch_size=batch_size, shuffle=False, drop_last=False)

    best_val = np.inf
    best_state = None
    no_improve = 0

    for epoch in range(1, max_epochs + 1):
        model.train()
        for xb, yb in train_loader:
            xb = xb.to(device); yb = yb.to(device)
            opt.zero_grad()
            pred = model(xb)
            loss = loss_fn(pred, yb)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()

        # validate
        yhat_val, y_val_np = eval_model(model, val_loader, device)
        if len(y_val_np):
            val_rmse = mean_squared_error(y_val_np, yhat_val)
        else:
            val_rmse = np.inf

        if val_rmse < best_val - 1e-6:
            best_val = val_rmse
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
            no_improve = 0
        else:
            no_improve += 1
        if no_improve >= early_stopping_patience:
            break

    if best_state is not None:
        model.load_state_dict(best_state)
    return model
```

```python
import modules.get_mrs_regime as gmr
# --- Regime feature builders (uses the horizon-aware pipeline we made earlier) ---

def _rename_probs_to_new_mapping(probs_df: pd.DataFrame, mapping_old_to_new: dict, prefix="mrs_p") -> pd.DataFrame:
    # map old regime columns -> new variance-ordered columns 0..k-1
    rename = {old: f"{prefix}{new}" for old, new in mapping_old_to_new.items()}
    out = probs_df.rename(columns=rename)
    # order columns as p0, p1, p2, ...
    new_cols = [f"{prefix}{i}" for i in range(len(mapping_old_to_new))]
    return out.reindex(columns=new_cols)

def _build_fold_regime_features(price: pd.Series,
                                train_end: pd.Timestamp,
                                te_start: pd.Timestamp,
                                te_end: pd.Timestamp,
                                horizon_days: int,
                                mrs_cfg=None):
    """
    Fit Markov regimes on train (<= train_end) and produce regime features for both
    train and test windows. Returns two DataFrames aligned to decision time t.
    """
    if mrs_cfg is None:
        mrs_cfg = gmr.MRSHorizonConfig()

    # Fit on full price but split internally at train_end (the function is leakage-safe)
    out = gmr.train_test_regimes_horizon(price, train_end, horizon_days, cfg=mrs_cfg)

    # --- TRAIN features (use PREDICTED probs for no in-sample leakage) ---
    probs_tr_pred = out.res_train.predicted_marginal_probabilities.copy()
    probs_tr_pred.index = out.y_train.index
    probs_tr_pred = _rename_probs_to_new_mapping(probs_tr_pred, out.mapping_old_to_new, prefix="mrs_p")

    # Hard label from predicted probs (not smoothed), with your persistence if you want:
    hard_tr = probs_tr_pred.idxmax(axis=1).str.replace("mrs_p", "", regex=False).astype(float).astype("Int64")
    hard_tr = hard_tr.rename("mrs_reg")
    # Optional persistence (same logic as before). Comment out if you prefer raw argmax:
    hard_tr = gmr.enforce_min_duration(hard_tr, min_bars=mrs_cfg.min_bars)

    feats_tr = pd.concat([probs_tr_pred, hard_tr], axis=1)
    feats_tr = feats_tr.loc[(feats_tr.index >= out.y_train.index.min()) & (feats_tr.index <= train_end)]

    # One-hot for label (as categorical; your pipeline will one-hot later if you keep it non-numeric)
    feats_tr["mrs_reg"] = feats_tr["mrs_reg"].astype("Int64")

    # --- TEST features (predicted probs from the test filter, already real-time) ---
    if out.probs_test_pred is not None:
        probs_te_pred = _rename_probs_to_new_mapping(out.probs_test_pred, out.mapping_old_to_new, prefix="mrs_p")
        hard_te = probs_te_pred.idxmax(axis=1).str.replace("mrs_p", "", regex=False).astype(float).astype("Int64")
        hard_te = hard_te.rename("mrs_reg")
        hard_te = gmr.enforce_min_duration(hard_te, min_bars=mrs_cfg.min_bars)
        feats_te = pd.concat([probs_te_pred, hard_te], axis=1)
        # Keep only the fold's test window
        feats_te = feats_te.loc[(feats_te.index >= te_start) & (feats_te.index <= te_end)]
    else:
        feats_te = pd.DataFrame(columns=["mrs_p0","mrs_p1","mrs_p2","mrs_reg"], index=pd.DatetimeIndex([], name=price.index.name))

    return feats_tr, feats_te
```

```python
def roll_forward_lstm(
    X: pd.DataFrame,         # features
    y: pd.Series,            # H-day target aligned to X.index
    price: pd.Series,
    train_years: int = 6,
    test_days: int = 126,    # half-year â‰ˆ 126 bd
    step_days: int = 126,
    embargo_days: int = 15,  # >= H business day buffer in calendar terms
    seq_len: int = 60,
    hidden: int = 64, layers: int = 1, dropout: float = 0.1,
    lr: float = 1e-3, weight_decay: float = 1e-4,
    batch_size: int = 128, max_epochs: int = 200, patience: int = 20,
    verbose: bool = True
):
    assert X.index.equals(y.index)
    idx = X.index
    n = len(idx)

    # split dtypes
    bool_cols = X.select_dtypes(include=["bool"]).columns.tolist()
    if bool_cols:
        X = X.copy()
        X[bool_cols] = X[bool_cols].astype("int8")
    num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = [c for c in X.columns if c not in num_cols]

    oos_pred = pd.Series(index=idx, dtype=float)
    oos_true = pd.Series(index=idx, dtype=float)

    start_pos = int(252 * train_years)
    fold = 0

    while True:
        tr_start_pos = max(0, start_pos - int(252 * train_years))
        tr_end_pos   = start_pos - 1
        te_start_pos = start_pos
        te_end_pos   = biz_offset_pos(te_start_pos, test_days, n) - 1
        if te_start_pos >= n or te_end_pos <= te_start_pos:
            break

        tr_start, tr_end = idx[tr_start_pos], idx[tr_end_pos]
        te_start, te_end = idx[te_start_pos], idx[te_end_pos]
        tr_mask = between_mask(idx, tr_start, tr_end)
        te_mask = between_mask(idx, te_start, te_end)

        # purge + embargo around test
        embargo = pd.Timedelta(days=embargo_days)
        pe_mask = (idx >= te_start - embargo) & (idx <= te_end + embargo)
        tr_mask = tr_mask & (~pe_mask)

        X_tr_raw, y_tr_raw = X.loc[tr_mask], y.loc[tr_mask]
        X_te_raw, y_te_raw = X.loc[te_mask], y.loc[te_mask]

        mrs_tr, mrs_te = _build_fold_regime_features(
            price=price,
            train_end=tr_end,
            te_start=te_start,
            te_end=te_end,
            horizon_days=horizon,
            mrs_cfg=gmr.MRSHorizonConfig(k_regimes=3, p_thresh=0.7, min_bars=4)
        )

        X_tr_raw = X_tr_raw.join(mrs_tr, how="left")
        X_te_raw = X_te_raw.join(mrs_te, how="left")

        # drop NaN target rows (no labels)
        mtr = y_tr_raw.notna(); X_tr_raw, y_tr_raw = X_tr_raw[mtr], y_tr_raw[mtr]
        mte = y_te_raw.notna(); X_te_raw, y_te_raw = X_te_raw[mte], y_te_raw[mte]

        # numeric pipeline fit on train only
        X_tr_num = X_tr_raw[num_cols] if num_cols else pd.DataFrame(index=X_tr_raw.index)
        X_te_num = X_te_raw[num_cols] if num_cols else pd.DataFrame(index=X_te_raw.index)

        if len(num_cols):
            imp, sc = fit_num_imputer_scaler(X_tr_num)
            X_tr_num_s = pd.DataFrame(transform_num_with(imp, sc, X_tr_num), index=X_tr_num.index, columns=num_cols)
            X_te_num_s = pd.DataFrame(transform_num_with(imp, sc, X_te_num), index=X_te_num.index, columns=num_cols)
        else:
            X_tr_num_s = pd.DataFrame(index=X_tr_raw.index)
            X_te_num_s = pd.DataFrame(index=X_te_raw.index)

        # categoricals one-hot (train-fitted)
        X_tr_cat = X_tr_raw[cat_cols] if cat_cols else pd.DataFrame(index=X_tr_raw.index)
        X_te_cat = X_te_raw[cat_cols] if cat_cols else pd.DataFrame(index=X_te_raw.index)
        X_tr_cat_oh, X_te_cat_oh = one_hot_train_test(X_tr_cat, X_te_cat)

        # combine processed numeric + cat
        X_tr_p = pd.concat([X_tr_num_s, X_tr_cat_oh], axis=1)
        X_te_p = pd.concat([X_te_num_s, X_te_cat_oh], axis=1)

        # time-ordered split inside TRAIN for early stopping (last 20% as val)
        cut_ts = X_tr_p.index[int(len(X_tr_p) * 0.8)] if len(X_tr_p) > 10 else X_tr_p.index[-1]
        tr_in_mask  = X_tr_p.index <= cut_ts
        val_in_mask = X_tr_p.index >  cut_ts

        # build sequences per slice
        Xtr_seq, ytr, _ = make_sequences_from_slice(X_tr_p, y_tr_raw, X_tr_p.index.min(), X_tr_p.index[tr_in_mask].max(), seq_len)
        Xva_seq, yva, _ = make_sequences_from_slice(X_tr_p, y_tr_raw, X_tr_p.index[val_in_mask].min() if val_in_mask.any() else X_tr_p.index.min(), X_tr_p.index.max(), seq_len)
        Xte_seq, yte, te_idx_seq = make_sequences_from_slice(X_te_p, y_te_raw, te_start, te_end, seq_len)

        # if no sequences, advance
        if Xte_seq.shape[0] == 0 or Xtr_seq.shape[0] == 0 or Xva_seq.shape[0] == 0:
            start_pos = biz_offset_pos(te_start_pos, step_days, n)
            continue

        # train one fold
        model = train_lstm_one_fold(
            Xtr_seq, ytr, Xva_seq, yva,
            hidden=hidden, layers=layers, dropout=dropout,
            lr=lr, weight_decay=weight_decay,
            batch_size=batch_size, max_epochs=max_epochs, early_stopping_patience=patience
        )

        # predict OOS
        te_loader = DataLoader(SeqDataset(Xte_seq, yte), batch_size=256, shuffle=False, drop_last=False)
        yhat, yt = eval_model(model, te_loader, "cuda" if torch.cuda.is_available() else "cpu")

        # store & metrics
        pred_s = pd.Series(yhat, index=te_idx_seq)
        true_s = pd.Series(yt,   index=te_idx_seq)
        oos_pred.loc[pred_s.index] = pred_s.values
        oos_true.loc[true_s.index] = true_s.values

        rmse = mean_squared_error(true_s, pred_s)
        mae  = mean_absolute_error(true_s, pred_s)
        ic   = spearmanr(true_s.values, pred_s.values, nan_policy="omit").correlation
        da   = (np.sign(true_s.values) == np.sign(pred_s.values)).mean()

        fold += 1
        if verbose:
            print(f"[Fold {fold}] Train {tr_start.date()}â†’{tr_end.date()} (n={len(X_tr_p)}) | "
                  f"Test {te_start.date()}â†’{te_end.date()} (n={len(pred_s)}) | "
                  f"RMSE={rmse:.5f} MAE={mae:.5f} DA={da*100:.2f}% IC={ic:.3f}")

        # advance
        start_pos = biz_offset_pos(te_start_pos, step_days, n)

    # overall stats
    valid = oos_true.notna() & oos_pred.notna()
    overall = {
        "n_oos_points": int(valid.sum()),
        "mse": float(mean_squared_error(oos_true[valid], oos_pred[valid])) if valid.any() else np.nan,
        "rmse": float(np.sqrt(mean_squared_error(oos_true[valid], oos_pred[valid]))) if valid.any() else np.nan,
        "mae": float(mean_absolute_error(oos_true[valid], oos_pred[valid])) if valid.any() else np.nan,
        "da": float((np.sign(oos_true[valid]) == np.sign(oos_pred[valid])).mean()) if valid.any() else np.nan,
        "ic": float(spearmanr(oos_true[valid].values, oos_pred[valid].values, nan_policy="omit").correlation) if valid.any() else np.nan,
        "seq_len": seq_len
    }
    return oos_pred, oos_true, overall
```
